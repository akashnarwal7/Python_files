# Multiple References in Python

```python
oldLi=[1,2,3,4]
newLi=oldLi

# on changing one refernce of list it will modify the original list and all refernces which point this list this change will also reflect to them.

newLi.append(5)
print("old List : ",oldLi)
print("new List : ",newLi)
```

# Slicing : way to get a subArray

#### list[start:end:step] 
#### start -> start index of the subarray
#### end   -> end is not included end-1 is end of the subarray
#### step  -> value by which index should increment

```python
li=[1,2,3,3,4]

li[1:3:1]
```

#### by default [start:end] if we don't give start by default it is zero 
#### by default end is last element in list and step by default is 1 

```python
li[1:3]
```

### li[:end] index after first column will be end
