```python
# create a set

set1={1,2,3}
type(set1)
```

```python
# we can also create a set.
# by converting any iterable into a set.
# by converting any sequence into a set.

set1=set([1,2,3])
set2=set((1,23,4))

# keys are only considered as elements.
set3=set({1:4,2:3})
print(set1)
print(set2)
set3
```

### Add

```python
set1={1,2,3}
print(set1)
set1.add(4)
set1
```

### Adding Multiple Elements

```python
set1={1,2,3}
print(set1)
set1.update([4,2,3])
set1
```

```python
set1.update((1,4,5,6))
print(set1)
```

```python
set1.update({'a':1,'b':2})
set1
```

### Remove

#### Remove Items

```python
# Remove value by an element.
# throws an error when ele is not found.
set1.remove('a')
set1
```

```python
# Remove value by an element
# no error
set1.discard('b')
set1
```

```python
#Deletes Randomly
set1.pop()
```

#### Remove or clear set

```python
# only content
set1.clear()
set1
```

```python
#Removes complete set.
del set1
set1
```

### copy

```python
# shallow copy means only content will be copied.
# deep copy means refernece is copied.

set1={1,2,3,4}
set2=set1.copy()
print(id(set1))
print(id(set2))
```

### Union ( | )

```python
# all elements in both sets but appear only once.
set1={1,2,3}
set2={3,4,5}
set.union(set1,set2)
```

```python
set1|set2
```

### InterSection ( & )

```python
# Only common elements
set1={1,2,3,4}
set2={3,4,5,6}
set.intersection(set1,set2)
```

```python
set1 & set2
```

### difference

```python
# only set1 elements which are uncommon
set1={1,2,3,4,5}
set2={4,5,6,7}
set.difference(set1)
```

### Symmetric Difference

```python
# all Uncommon elements in both sets
set1={1,2,3,4,5}
set2={4,5,6,7}
set.symmetric_difference(set1,set2)
```

### issubset

```python
set1={1,2}
set2={1,2,3,4}
set1.issubset(set2)
```

### issuperset

```python
set2.issuperset(set1)
```
