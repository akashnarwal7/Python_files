## Series

### series are also similar as Dictionaries 
### But series has 2 indices 
### 1. Position based Index which is 0,1,2,....
### 2. Label based Index this is optional we can give labels as anything and use them as index.

```python
import pandas as pd

series1=pd.Series([10,20,30,40,50])
print("Position Based Indexing :",series1)
```

## Label Based Indexing 
### Give all indices or labels as this 
### index=[list of labels or indices]

```python
series2=pd.Series([10,20,30,40,50],index=["Akash",'Praveen','sindhu','lilly','andy'])
print("Label Based Indexing")
series2
```

# Read Any type of Data and store it in DataFrames.

## Lists to DataFrames

```python
Data=pd.DataFrame([
    [10, 'akash', 'cse'],
    [40,'andy','Finance'],
    [50,'lilly','mech']
])
Data
```

```python
# we can also Label cols
Data=pd.DataFrame([
    [10, 'akash', 'cse'],
    [40,'andy','Finance'],
    [50,'lilly','mech']
],columns=["No", "Name","Branch"])
Data
```

## Dicitionaries to DataFrames.

```python
dic={'a':[1],'b':[2],'c':[3]}
data=pd.DataFrame(dic)
data
```

```python
dic={'number':[1,2,3,4],'name':['akash','Andy','lilly','jean']}
data=pd.DataFrame(dic)
data
```
