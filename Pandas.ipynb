# Data Filtering

### Data Filetring Means Applying Filters or Applying Conditions to Data.
### we can Filter Data using 
### 1.Conditional Indexing
### 2.Query() Method

## Using Conditional Indexing

```python
Data
```

```python
# Apply Conditions as Index.
Data[ Data['PetalWidthCm'] >1.8]
```

```python
## Apply Multiple Conditions.
## use & as logical and but use (con1) & (con2)
## use  parantheses for condiitons. 

Data[ (Data['PetalWidthCm'] ==1.8) & (Data['SepalWidthCm']>3.0)]
```

```python
Data[(Data['PetalWidthCm'] ==1.8)  | (Data['SepalWidthCm']>3.0)]
```

# Using Query()

# syntax :
# dataframe.query(" condition as query ")




### we can use columns with their column names no need of indexing of columns is required 
### for ex : in conditional indexing to access species we need to use :
### Data["Species"]
### but in query we can address it with the column names only

```python
Data.query(" Species=='Iris-virginica'")
```

```python
# to use && ,|| ,! donot use them
# rather use and ,not,or
Data.query(" Species != 'Iris-virginica' and  SepalLengthCm >5.8 ")
```

```python
Data.query(" not (Species == 'Iris-virginica' and  SepalLengthCm <3) ")
```

## use @variable to refer to python variable

```python
flowerType='Iris-virginica'
Data.query("Species== @flowerType ")
```

## Normally we can use >=,<=,==,!=,>,<

### chained comparisons

```python
Data.query("40<Id<=70")
```

# Solution : .copy()

```python
copy=data['name'].copy()
copy
```

```python
copy.loc['Row2']='praveen'
print("Only Change in copy")
print(copy)
print()
print("data")
print()
print(data)
```
