# Tuples

#### Tuples are also sequence of items or even sequences
#### . It must have  commas
#### . Parantheses must be optional
#### . Tuples are immutable we cannot change tuples

### Creating Tuples

### Tuples can be cretaed by assigning them as a literal

```python
t1=()
#empty literal
print(type(t1))
```

```python
t1=(1,2,3)
print(type(t1))
```

```python
# comma is must in tuples so () and atleast one comma must be present 
# so when we have one element in tuple comma must be there at the end.

t1=(1)
type(t1)
```

```python
t1=(1,)
print(type(t1))
```

```python
# Without Parantheses
t1=1,2,3
print(type(t1))
```

# Indexing , Slicing

### Tuples on indexing,slicing are similar to lists

# changing tuples

# we can't change tuple as it is immutable

```python
t1=(21,22,3)
t1[1]=3
```

### But we can change its mutable item.

```python
t1=(1,2,['a','b'])
t1[-1][0]='c'
print(t1)
# here we are not making any change in tuple 
# reason the item in tuple is not list ['a','b'] but the address of list so changing list won't change its address so mutable items can be modifeid.
```

# Concatenation of Tuples

### Using + operator

```python
t1=('a','b')+('c','d')+('e','f')
print(t1)
```

### Using * operator

```python
# same tuple can be concatenated mupltiple times.
t1=('a','b')*4
t1
```

# Deletion of tuple

```python
# we cannot delete particular item as it is immutable
# we cannot replace items ,modify items in tuple
del t1
t1
```
