# Changing characters in string

### Changing characters in string is not possible as it is immutable
### but we can reassign a new string
### Using replace function we get a new string but not modifying old string .

```python
s='akas'
s='a'
s
```

### Deleting a item in string is not possible as it is immutable

```python
s='abs'
del s[0]
```

### Deleting a string is possible

```python
s='abs'
del s
s
```
